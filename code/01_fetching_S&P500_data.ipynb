# CAPSTONE Part 1: Fetching S&P500 Historical Stock Price Data

# 1. Importing Libraries

```python
# Load libraries
import pandas as pd
import yfinance as yf
import warnings
warnings.filterwarnings('ignore')
```

# 2. Fetching the data for all the stocks of S&P500 from the start of 2000 to September 2023

```python
# Define the URL of the Wikipedia page containing the S&P 500 stock list
snp500url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
# Use pandas to scrape the tables from the Wikipedia page and store them in a list
data_table = pd.read_html(snp500url)
```

```python
# Extract the 'Symbol' column from the DataFrame data_table at index 0, excluding the first row and convert it into a list of ticker symbols
tickers = data_table[0][1:]['Symbol'].tolist()
# Print the list of ticker symbols
print(tickers)
# Print the total number of companies in the list of ticker symbols
print("\nTotal number of companies are:", len(tickers))
```

```python
# Define the start and end dates for the historical stock price data
start = '2000-01-01'
end = '2023-09-10'

# Initialize an empty DataFrame to store the 'Adj Close' columns for multiple stocks
adj_close_data = pd.DataFrame()

# Loop through each ticker symbol and download historical stock price data
for ticker in tickers:
    # Download historical stock price data for the current ticker symbol
    data = yf.download(ticker, start , end)
    # Extract the 'Adj Close' column from the downloaded data
    data_adj_close = data[['Adj Close']]
    # Rename the 'Adj Close' column to match the current ticker symbol
    data_adj_close.rename(columns ={'Adj Close': ticker}, inplace = True)
    # Concatenate the 'Adj Close' column to the existing DataFrame
    adj_close_data = pd.concat([adj_close_data, data_adj_close], axis=1)
```

```python
# The resulting adj_close_data DataFrame will have date as the index and 'Adj Close' columns for each ticker
adj_close_data.head()
```

```python
# Change index datetime format to remove hours:minutes:seconds
adj_close_data.index = pd.to_datetime(adj_close_data.index, format='%Y-%m-%d')
```

```python
# Check that index has the correct datatype
adj_close_data.index
```

```python
# Check that index is in the required format
adj_close_data.head()
```

```python
# Save the data to csv format 
adj_close_data.to_csv("../data/S&P500.csv")
```
