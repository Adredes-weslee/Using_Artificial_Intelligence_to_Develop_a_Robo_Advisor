**Features:**

**Target Variable:** 

1. For the problem statement to be solved using supervised ML, the target variable is risk tolerance. This is first calculated from a ratio of risky to risk + risk-free assets. Selected demographic, financial and behavioral attributes of an individual are then used as the feature variables to predict the risk tolerance
2. Reinforcement learning is neither supervised nor unsupervised as it does not require labeled data or a training set. It relies on the ability to monitor the response to the actions of the learning agent. For the valued-based DQN, the reward is the sharpe ratio of the portfolio. Maximizing this reward will lead to portfolio allocation weights that maximize the return of the portfolio for a given level of risk

**Evaluation Metrics:**  

1. Evaluate the supervised ML models based on their R2 and RMSE values
2. For the value-based DQN, the loss function minimized by the ANN through gradient descent is the squared difference between the DQN’s estimate of the target and its estimate of the Q-value of the current state-action pair, Q(s,a:θ). We don't directly evaluate the value of the loss function, but since the Q-value is the expected reward for the state-action pair following a policy π, when the algorithm iteratively converges to the optimal Q-value, it learns an optimal policy which is how to act to maximize the return/reward in every state, which is to maximize the sharpe ratio of the portfolio. By maximizing the sharpe ratio of the portfolio, the RL agent also dynamically (automatically) change the portfolio allocation weights

**Significance:** 

The risk tolerance of an investor is one of the most important inputs to the portfolio allocation and rebalancing steps of the portfolio management process. There are a wide variety of risk profiling tools that take varied approaches to understanding the risk tolerance of an investor. Most of these approaches include qualitative judgment and involve significant manual effort. In most of the cases, the risk tolerance of an investor is decided based on a risk tolerance questionnaire. Several studies have shown that these risk tolerance questionnaires are prone to error, as investors suffer from behavioral biases and are poor judges of their own risk perception, especially during stressed markets. Also, given that these questionnaires must be manually completed by investors, they eliminate the possibility of automating the entire investment management process. The overall purpose is to demonstrate the automation of the manual steps in the portfolio management process with the help of machine learning. This can prove to be immensely useful for a robo-advisor.

Once the risk tolerance of an investor has been determined, the next logical step would be to select assets. Modern portfolio theory suggests that optimal (or efficient) portfolios can be constructed by meaningfully allocating individual asset weights in the correct proportion for a portfolio so as to maximize the sharpe ratio which is also the greatest expected return for a given level of risk (i.e., standard deviation), or, equivalently, the lowest risk for a given expected return. Therefore the portfolio allocation process should determine the optimal weights for any given portfolio for our investor with their own unique risk tolerance value, so as to give the best trade-off between expected return and risk.

How reinforcement learning comes in is that reinforcement learning algorithms, with the ability to decide the policy on their own, are strong models for performing portfolio allocation in an automated manner, without the need for continuous supervision in rebalancing portfolio weights. Automation of the manual steps involved in portfolio allocation can prove to be immensely useful, specifically for robo-advisors. In an RL-based framework, portfolio allocation is treated not just as a one-step optimization problem but as continuous control of the portfolio with delayed rewards. This means making ongoing adjustments based on real-time or frequent data updates. By moving from discrete optimal allocation to continuous control territory, and in the environment of a continuously changing market, RL algorithms can be leveraged to solve complex and dynamic portfolio allocation problems.

Second, while human advice is costly, RAs charge much lower fees. Third, they can offer investment services for much lower investment amounts than traditional advisors. By providing low-fee, online and rules-based investment advice, robo-advisory can improve personal investing.

RAs have grown over time with respect to their capabilities and are now the furthest that digitalization has come in terms of investment management for retail clients. On a separate note, but related to AI in finance in general, financial theories are normative, being based on a lot of unrealistic assumptions that have been proven to be false. # 6. Preparing the target variable (Risk_tolerance)

In the first step, we prepare the predicted variable, which is the true risk tolerance.
The steps to compute the true risk tolerance are as follows:

1. Compute the risky assets and the risk-free assets for all the individuals in the survey
data. Risky and risk-free assets are defined as follows:
    - **Risky assets**: Stocks, bonds, directly held pooled investment funds and various others (see Data Dictionary above in Section 5)
    - **Risk-free assets**: Transaction accounts, money market deposit and money market mutual fund accounts, checking accounts, savings accounts, call accounts, certificates of deposit, prepaid card accounts, savings bonds, cash value of whole life insurance

2. Take the ratio of risky assets to total assets (where total assets is the sum of risky and risk-free assets) of an individual and consider that as a measure of the individual’s risk tolerance

# 7. Modeling

## 7.1 Train/Test Split

```python
# Set seed
seed = 42
# Create target matrix 
y = working_df['Risk_tolerance']
# Create features matrix
X = working_df.drop(labels = ['Risk_tolerance'], axis = 1)
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = seed)
```

## 7.2 Function to Create Sequential Neural Network Using Keras

```python
# Create DNN model using keras
def baseline_model():
    # Create model
    model = Sequential()

    # Input layer with 32 neurons, input shape determined by X_train
    model.add(Dense(32, input_dim=X_train.shape[1], kernel_initializer='normal', activation='relu'))
    model.add(Dropout(0.5))  # Dropout layer to prevent overfitting

    # Hidden layers with various numbers of neurons
    model.add(Dense(16, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(8, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(4, activation='relu'))
    model.add(Dropout(0.5))

    # Final output layer with 1 neuron (regression task)
    model.add(Dense(1))

    # Compile model with custom loss function and optimizer
    model.compile(loss='mse', optimizer='adam', metrics=['mse'])

    return model
```

## 7.3 Instantiating Regressor Classes

```python
# Choose number of folds for k-folds cross validation
num_folds = 10
# Choose scoring metric for regression
scoring = 'neg_root_mean_squared_error'

# Instantiating models
models = []
models.append(('LR', LinearRegression()))
models.append(('LASSO', Lasso()))
models.append(('EN', ElasticNet()))
models.append(('KNN', KNeighborsRegressor()))
models.append(('CART', DecisionTreeRegressor()))
models.append(('SVR', SVR()))

# Neural Network
models.append(('MLP', MLPRegressor()))

## Ensemble Models 

# Boosting methods
models.append(('ABR', AdaBoostRegressor()))
models.append(('GBR', GradientBoostingRegressor()))

# Bagging methods
models.append(('RFR', RandomForestRegressor()))
models.append(('ETR', ExtraTreesRegressor()))

# Deep Learning Regressor
models.append(('DNN', KerasRegressor(build_fn=baseline_model, epochs=50, batch_size=100, verbose=0))) 
```

## 7.4 Training the Regression Models

```python
# Cross validation mean and std RMSE
results = []
names = []
for name, model in models:
    kfold = KFold(n_splits=num_folds, shuffle=True, random_state=seed)
    #converted mean square error to positive. The lower the beter
    cv_results = -1* cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
# Train/test RMSE and R2 values
for name, model in models:
    model.fit(X_train, y_train)
    r2_train = model.score(X_train, y_train)
    r2_test = model.score(X_test, y_test)
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    train_rmse = mean_squared_error(y_train, y_pred_train, squared = False)
    test_rmse = mean_squared_error(y_test, y_pred_test, squared = False)
    print(f'For {name}, the train RMSE/R2 is {np.round(train_rmse, 5)}/{np.round(r2_train, 5)}, the test RMSE/R2 is {np.round(test_rmse, 5)}/{np.round(r2_test, 5)}')
```

## 7.5 Regression Models Cross-validated RMSE Comparison

```python
# Compare models cross-validated RMSE
fig = plt.figure()
fig.suptitle('Regression Models Cross-validated RMSE Comparison', fontsize=30)
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
fig.set_size_inches(15,8)
plt.xlabel('Models', fontsize=20)
plt.ylabel('RMSE', fontsize=20)
plt.show();

# plt.savefig('../images/Model_crossvalidation_comparison.png', dpi=1000);
```

The non linear models perform better than the linear models, which means that a non linear relationship must exist between risk tolerance and the different variables use to predict risk tolerance. Given that the extra trees regressor is the best model based on the mean of the cross-validated RMSE score, we shall select it for hyperparameter tuning via GridSearch.
